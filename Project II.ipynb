```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Importing Datasets

```python
matches = pd.read_csv('matches.csv')
players = pd.read_csv('Players.csv')
indianPlayers = pd.read_csv("Indian Players.csv")
indianPlayersWithRuns = pd.read_csv("Indian Players with runs and strikerate.csv")
foreignPlayers = pd.read_csv('Foreign Playes.csv')
foreignPlayersWithRuns = pd.read_csv('Foreign Playes with runs and strikerate.csv')
```

# Ploting graphs

```python
matchWin = matches['winner'].value_counts()
matchWin
```

```python
tosswin = matches['toss_winner'].value_counts()  
tosswin
```

```python
Teams_Of_Win = pd.Series(['Mumbai Indians','Chennai Super Kings','Kolkata Knight Riders','Royal Challengers Bangalore','Kings XI Punjab','Rajasthan Royals','Delhi Daredevils','Sunrisers Hyderabad','Deccan Chargers','Gujarat Lions','Pune Warriors','Delhi Capitals','Kochi Tuskers Kerala','Rising Pune Supergiants'])
count_Of_Win = pd.Series([109,98,92,82,81,75,66,56,29,13,12,9,6,5])
```

```python
Teams_Of_Toss = pd.Series(['Mumbai Indians','Kolkata Knight Riders','Chennai Super Kings','Kings XI Punjab','Delhi Daredevils','Rajasthan Royals','Royal Challengers Bangalore','Sunrisers Hyderabad','Deccan Chargers','Pune Warriors','Gujarat Lions','Delhi Capitals','Kochi Tuskers Kerala','Rising Pune Supergiants'])
count_Of_toss = pd.Series([94,91,89,80,79,77,76,44,43,20,15,9,8,8])
```

```python
TeamByWin = pd.DataFrame({"Teams":Teams_Of_Win, "No of wins":count_Of_Win})
```

```python
TeamByWin
```

```python
TeamByToss = pd.DataFrame({"Teams":Teams_Of_Toss, "No of wins":count_Of_toss})
```

```python
TeamByToss
```

```python
TeamByWin.to_csv("Count_of_teams_wins.csv")
```

```python
TeamByToss.to_csv("Count_of_teams_wins_by_toss.csv")
```

```python
# Count of match winner 
fig,ax=plt.subplots(figsize=(18,8))
ax.barh(TeamByWin['Teams'],TeamByWin['No of wins']);
ax.set(title="Count of match winner",
      xlabel="Wins")
```

```python
# Count of toss winner
plt.style.use('seaborn-notebook')
TeamByToss.plot(x="Teams", y="No of wins", kind="bar", ylabel="No of toss wins")
```

```python
choice = ['field','bat']
countofchoice = [445,288]
```

```python
# Count of choice of batting or field
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
mycolors = ["deepskyblue", "lightblue"]
ax.set_title("Toss decision by toss winner")
ax.pie(countofchoice, labels = choice,autopct='%1.2f%%', shadow=True, colors = mycolors)
plt.show()
```
